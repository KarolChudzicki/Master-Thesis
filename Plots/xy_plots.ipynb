```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Load CSV
df = pd.read_csv("follow/xy_2025-08-09_21-42-27.csv")  

# Convert raw timestamp to relative time
df["relative_time"] = df["time"] - df["time"].iloc[0]

# Scale factor
scale = 1.8

# Create figure and gridspec layout with bigger figure size
fig = plt.figure(figsize=(30*scale, 18*scale))  # Scale figure size

gs = gridspec.GridSpec(3, 3, figure=fig)

# --- Top row ---

ax_x = fig.add_subplot(gs[0, 0])
ax_x.plot(df["relative_time"], df["x_camera"], label="X Meas.", marker="D", markersize=8*scale)
ax_x.plot(df["relative_time"], df["x_pred"], label="X Pred.", marker="x", markersize=14*scale)
ax_x.plot(df["relative_time"], df["x_final"], label="X Final", marker="s", markersize=6*scale)
ax_x.set_title("X Distance [m]", fontsize=22*scale, fontweight='bold')
ax_x.set_xlabel("Time [s]", fontsize=20*scale)
ax_x.set_ylabel("Distance [m]", fontsize=20*scale)
ax_x.grid(True)
ax_x.legend(fontsize=18*scale)
ax_x.tick_params(axis='both', labelsize=18*scale)

ax_y = fig.add_subplot(gs[0, 1])
ax_y.plot(df["relative_time"], df["y_camera"], label="Y Meas.", marker="D", markersize=8*scale)
ax_y.plot(df["relative_time"], df["y_pred"], label="Y Pred.", marker="x", markersize=14*scale)
ax_y.plot(df["relative_time"], df["y_final"], label="Y Final", marker="s", markersize=6*scale)
ax_y.set_title("Y Distance [m]", fontsize=22*scale, fontweight='bold')
ax_y.set_xlabel("Time [s]", fontsize=20*scale)
ax_y.set_ylabel("Distance [m]", fontsize=20*scale)
ax_y.grid(True)
ax_y.legend(fontsize=18*scale)
ax_y.tick_params(axis='both', labelsize=18*scale)

# --- Second row ---

ax_xspeed = fig.add_subplot(gs[1, 0])
ax_xspeed.plot(df["relative_time"], df["x_speed"], color="blue", marker="o", linestyle="-", markersize=10*scale)
ax_xspeed.set_title("X Speed [m/s]", fontsize=22*scale, fontweight='bold')
ax_xspeed.set_xlabel("Time [s]", fontsize=20*scale)
ax_xspeed.set_ylabel("X Speed [m/s]", fontsize=20*scale)
ax_xspeed.grid(True)
ax_xspeed.tick_params(axis='both', labelsize=18*scale)

ax_yspeed = fig.add_subplot(gs[1, 1])
ax_yspeed.plot(df["relative_time"], df["y_speed"], color="green", marker="o", linestyle="-", markersize=10*scale)
ax_yspeed.set_title("Y Speed [m/s]", fontsize=22*scale, fontweight='bold')
ax_yspeed.set_xlabel("Time [s]", fontsize=20*scale)
ax_yspeed.set_ylabel("Y Speed [m/s]", fontsize=20*scale)
ax_yspeed.grid(True)
ax_yspeed.tick_params(axis='both', labelsize=18*scale)

# --- Third row: Alpha plot spanning two columns ---

ax_alpha = fig.add_subplot(gs[2, 0:2])
ax_alpha.plot(df["relative_time"], df["alpha"], color="purple", marker="o", linestyle="-", markersize=14*scale)
ax_alpha.set_title("Alpha", fontsize=22*scale, fontweight='bold')
ax_alpha.set_xlabel("Time [s]", fontsize=20*scale)
ax_alpha.set_ylabel("Alpha", fontsize=20*scale)
ax_alpha.grid(True)
ax_alpha.tick_params(axis='both', labelsize=18*scale)

plt.subplots_adjust()
plt.tight_layout(h_pad=10.0, w_pad=10.0)
plt.show()
```
