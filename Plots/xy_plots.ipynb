```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("follow/xy_2025-07-11_14-44-03.csv")

# Optional: convert raw timestamp to relative time
df["relative_time"] = df["time"] - df["time"].iloc[0]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df["relative_time"], df["x_speed"], label="X speed")
plt.plot(df["relative_time"], df["y_speed"], label="Y speed")


plt.xlabel("Time [s]")
plt.ylabel("Values")
plt.title("Descend")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
