```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("follow/xy_2025-07-21_11-07-11.csv")

# Optional: convert raw timestamp to relative time
df["relative_time"] = df["time"] - df["time"].iloc[0]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df["relative_time"], df["x_speed"], label="X speed")
plt.plot(df["relative_time"], df["y_speed"], label="Y speed")


plt.xlabel("Time [s]")
plt.ylabel("Values")
plt.title("Descend")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSVs
df1 = pd.read_csv("plots_for_thesis/old_alpha.csv")      # contains alpha
df2 = pd.read_csv("plots_for_thesis/for_old_alpha.csv")  # contains x_speed and y_speed

# Convert raw timestamp to relative time
df1["relative_time"] = df1["time"] - df1["time"].iloc[0]
df2["relative_time"] = df2["time"] - df2["time"].iloc[0]

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 4), sharex=True)

# X Speed
axes[0].plot(df2["relative_time"], df2["x_speed"], color="blue", marker='o', linestyle='-')
axes[0].set_title("X Speed [m/s]")
axes[0].set_xlabel("Time [s]")
axes[0].set_ylabel("X Speed")
axes[0].grid(True)

# Y Speed
axes[1].plot(df2["relative_time"], df2["y_speed"], color="green", marker='o', linestyle='-')
axes[1].set_title("Y Speed [m/s]")
axes[1].set_xlabel("Time [s]")
axes[1].set_ylabel("Y Speed")
axes[1].grid(True)

# Alpha
axes[2].plot(df1["relative_time"], df1["alpha"], color="purple", marker='o', linestyle='-')
axes[2].set_title("Alpha")
axes[2].set_xlabel("Time [s]")
axes[2].set_ylabel("Alpha")
axes[2].grid(True)

# Layout
plt.tight_layout()
plt.show()
```
