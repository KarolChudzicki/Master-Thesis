```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("follow/xy_2025-08-06_15-37-51.csv")  
# Make sure it includes: time, x_camera, x_pred, x_final, y_camera, y_pred, y_final, x_speed, y_speed, alpha

# Convert raw timestamp to relative time
df["relative_time"] = df["time"] - df["time"].iloc[0]

# Create subplots: 2 rows, first row with 2 plots, second row with 3 plots
fig = plt.subplots(2, 3, figsize=(18, 8))[0]  # axes returned later
axes = plt.subplots(2, 3, figsize=(18, 8))[1]
plt.close()  # close extra figure from the double call
fig, axes = plt.subplots(2, 3, figsize=(18, 8))

# --- Top Row ---

# X variables
axes[0,0].plot(df["relative_time"], df["x_camera"], label="X Camera", marker="o")
axes[0,0].plot(df["relative_time"], df["x_pred"], label="X Predicted", marker="x")
axes[0,0].plot(df["relative_time"], df["x_final"], label="X Final", marker="s")
axes[0,0].set_title("X Distances")
axes[0,0].set_xlabel("Time [s]")
axes[0,0].set_ylabel("Distance [m]")
axes[0,0].grid(True)
axes[0,0].legend()

# Y variables
axes[0,1].plot(df["relative_time"], df["y_camera"], label="Y Camera", marker="o")
axes[0,1].plot(df["relative_time"], df["y_pred"], label="Y Predicted", marker="x")
axes[0,1].plot(df["relative_time"], df["y_final"], label="Y Final", marker="s")
axes[0,1].set_title("Y Distances")
axes[0,1].set_xlabel("Time [s]")
axes[0,1].set_ylabel("Distance [m]")
axes[0,1].grid(True)
axes[0,1].legend()

# Remove unused top-right plot (since only 2 on top row)
fig.delaxes(axes[0,2])

# --- Bottom Row ---

# X Speed
axes[1,0].plot(df["relative_time"], df["x_speed"], color="blue", marker="o", linestyle="-")
axes[1,0].set_title("X Speed [m/s]")
axes[1,0].set_xlabel("Time [s]")
axes[1,0].set_ylabel("X Speed")
axes[1,0].grid(True)

# Y Speed
axes[1,1].plot(df["relative_time"], df["y_speed"], color="green", marker="o", linestyle="-")
axes[1,1].set_title("Y Speed [m/s]")
axes[1,1].set_xlabel("Time [s]")
axes[1,1].set_ylabel("Y Speed")
axes[1,1].grid(True)

# Alpha
axes[1,2].plot(df["relative_time"], df["alpha"], color="purple", marker="o", linestyle="-")
axes[1,2].set_title("Alpha")
axes[1,2].set_xlabel("Time [s]")
axes[1,2].set_ylabel("Alpha")
axes[1,2].grid(True)

# Layout
plt.tight_layout()
plt.show()
```
